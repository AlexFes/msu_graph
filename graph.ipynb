# Map data to graph

```python
import xlrd
import csv
import pandas as pd
```

## Read data

```python
# Parse txt file and drop extra column
with open('data.txt') as file:
    reader = csv.reader(file, delimiter='\t')
    
    with open('miRBase.csv', 'w', newline='') as target_file:
        for row in reader:
            writer = csv.writer(target_file)
            
            if len(row) > 20:
                writer.writerow(row[:-1])
            else: 
                writer.writerow(row)
```

```python
with open('bc/data_good.txt') as file:
    reader = csv.reader(file, delimiter='\t')    
    with open('processed/good.csv', 'w') as target_file:
        for row in reader:
            writer = csv.writer(target_file)
            writer.writerow(row)

with open('bc/Total_old.txt') as file:
    reader = csv.reader(file, delimiter='\t')    
    with open('processed/total.csv', 'w') as target_file:
        for row in reader:
            writer = csv.writer(target_file)
            writer.writerow(row)
```

```python
# Parse dat file
with open('miRNA.dat') as file:
    reader = csv.reader(file)
    arr = []
    
    for row in reader:
        arr.append(row[0].split())

# Get only id and pointer rows        
arr = list(filter(lambda x: x[0] == 'ID' or (x[0] == 'FT' and 'product' in x[1]), arr))
arr = list(map(lambda x: ['FT', x[1][10:-1]] if ('FT' in x[0]) else x[:2], arr))

with open('processed/miRNA.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["precursor", "miRNA"])
    id = ''
    
    for row in arr:
        if row[0] == 'ID':
            id = row[1]
        else:
            writer.writerow([id, row[1]])
```

```python
miRBase = pd.read_csv("./miRBase.csv")
miRTarBase = pd.read_excel("./miRTarBase_MTI.xlsx")
miRNA = pd.read_csv('processed/miRNA.csv')
```

## Preprocess

```python
miRTarBase = miRTarBase.loc[miRTarBase['Species (miRNA)'] == 'Homo sapiens']
miRTarBase = miRTarBase[['miRNA', 'Target Gene']]
miRBase = miRBase[['precursor', 'host_symbol']]
```

```python
miRTarBase = miRTarBase.rename(columns={"Target Gene": "target"})
miRBase = miRBase.rename(columns={"host_symbol": "host"})
```

```python
miRTarBase = miRTarBase.drop_duplicates().dropna()
miRBase = miRBase.drop_duplicates().dropna()
miRNA = miRNA.drop_duplicates().dropna()
```

## Write to file

```python
miRTarBase.to_csv('processed/miRTarBase.csv', encoding='utf-8', index=False)
miRBase.to_csv('processed/miRBase.csv', encoding='utf-8', index=False)
miRNA.to_csv('processed/miRNA.csv', encoding='utf-8', index=False)
```
